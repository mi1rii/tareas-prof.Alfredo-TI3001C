#### Por cada filtro agregar un archivo .csv con el nombre de cada filtro y describir cuales fueron las variables filtradas dentro del script

```python
# Leer el CSV
import pandas as pd

df = pd.read_csv('Datos_Facturación.csv')
```

## 1. CVE_CLPV (valores: 1000 a 2000)

```python
cve_clpv = df[(df['CVE_CLPV'] >= 1000) & (df['CVE_CLPV'] <= 2000)]
cve_clpv.to_csv('filtro_CVE_CLPV_1000_2000.csv', index=False)
cve_clpv.head()
```

## 2. CVE_VEND (Todas las claves excepto “5” y “4”)

```python
cve_vend = df[~df['CVE_VEND'].isin([4,5])]
cve_vend.to_csv('filtro_CVE_VEND_excluye_4_5.csv', index=False)
cve_vend.head()
```

## 3. FECHA_ENT (Fechas de “28/02/2022”)

```python
fecha_ent = df[df['FECHA_ENT'] == '2019-02-28']
fecha_ent.to_csv('filtro_FECHA_ENT_2019-02-28.csv', index=False)
fecha_ent.head()
```

## 4. CAN_TOT (Cantidades menores a 5951.7) o STATUS (“E”)

```python
can_tot = df[(df['CAN_TOT'] < 5951.7) | (df['STATUS'] == 'E')]
can_tot.to_csv('filtro_CAN_TOT_menor_5951.7_o_STATUS_E.csv', index=False)
can_tot.head()
```

## 5. Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT

```python
filtro5 = df[['CVE_DOC', 'FECHA_ENT', 'FECHA_VEN', 'CAN_TOT']]
filtro5.to_csv('filtro_columnas_CVE_DOC_FECHA_ENT_FECHA_VEN_CAN_TOT.csv', index=False)
filtro5.head()
```

## 6. Solo las filas de 7001-7099

```python
filtro6 = df.iloc[7000:7099]
filtro6.to_csv('filtro_filas_7001_7099.csv', index=False)
filtro6.head()
```

## 7. Index= CVE_VEND (valores: 1, 2) (columna: FECHAELAB)

```python
filtro7 = df[df['CVE_VEND'].isin([1,2])][['FECHAELAB']]
filtro7.to_csv('filtro_CVE_VEND_1_2_FECHAELAB.csv', index=False)
filtro7.head()
```
